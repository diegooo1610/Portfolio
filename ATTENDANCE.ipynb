```python
df['Attendance'] = df['Attendance'].str.replace(',', '').astype(int)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python


# Assuming df is already loaded and df['Date'] is datetime

# Match Stage
df['Match_Stage'] = pd.cut(df.index, bins=[-1, 47, 55, 59, 61, 62, 63],
                           labels=['Group Stage', 'Round of 16', 'Quarterfinals', 'Semifinals', 'Third Place', 'Final'])

# Day of Week
df['Weekday'] = df['Date'].dt.day_name()

# Country (for this dataset, all in Qatar)
df['Country'] = 'Qatar'
```

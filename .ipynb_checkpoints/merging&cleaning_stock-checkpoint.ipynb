```python
from datetime import datetime, timedelta
import yfinance as yf
import pandas as pd

# Tickers list
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NVDA", "META", "JPM", "V", "JNJ"]

# Dynamic date range: 12 months back to today
end_date = datetime.today().replace(day=1)  # Start of current month (ensures full last month is included)
start_date = end_date - timedelta(days=365)

# Download data
data = yf.download(tickers, start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'), auto_adjust=True)

# Extract adjusted close prices
close_prices = data['Close'] if isinstance(data.columns, pd.MultiIndex) else data

# Daily returns
daily_returns = close_prices.pct_change()

# Monthly stats
monthly_stats = pd.DataFrame()
for ticker in tickers:
    grouped = daily_returns[ticker].resample('ME')  # Use 'ME' for Month-End
    stats = pd.DataFrame({
        'Mean_Return': grouped.mean(),
        'Min_Return': grouped.min(),
        'Max_Return': grouped.max(),
        'Volatility': grouped.std()
    })
    stats['Ticker'] = ticker
    monthly_stats = pd.concat([monthly_stats, stats])

monthly_stats.reset_index(inplace=True)
monthly_stats.rename(columns={'Date': 'Month'}, inplace=True)

# Save to CSV
monthly_stats.to_csv("monthly_stock_statistics_dynamic.csv", index=False)
print("✅ Saved updated data to 'monthly_stock_statistics_dynamic.csv'")

```

```python
# Ensure correct data types
df['Month'] = pd.to_datetime(df['Month'])
df['Ticker'] = df['Ticker'].astype(str)

# Drop rows with missing values
df.dropna(inplace=True)

# Confirm it's clean
print(df.info())
print(df.describe())
```

```python
df['Month_Num'] = df['Month'].dt.month
df['Year'] = df['Month'].dt.year
```

```python
df['Volatility_Level'] = pd.cut(
    df['Volatility'],
    bins=[0, 0.02, 0.05, 1],
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Prev_Month_Return'] = df.groupby('Ticker')['Mean_Return'].shift(1)
```

```python
import matplotlib.pyplot as plt

# Get the year range from your dataset
start_year = df['Year'].min()
end_year = df['Year'].max()

# Group and plot
df.groupby('Month_Num')['Volatility'].mean().plot(kind='bar', figsize=(8, 4))
plt.title(f"📉 Average Monthly Volatility ({start_year}–{end_year})")
plt.xlabel("Month")
plt.ylabel("Volatility")
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Month_Num', y='Volatility', hue='Ticker', marker='o')
plt.title(f"Monthly Volatility by Ticker ({start_year}–{end_year})")
plt.xlabel("Month")
plt.ylabel("Volatility")
plt.grid(True)
plt.tight_layout()
plt.show()
```
